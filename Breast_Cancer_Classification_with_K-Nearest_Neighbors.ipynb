```python
from matplotlib import pyplot as plt
from joblib import dump, load
import pandas as pd
import numpy as np
from sklearn import datasets
from sklearn.metrics import accuracy_score
```

```python
cancer = datasets.load_breast_cancer()
```

```python
# Extracting Attributes / Features
X = cancer.data

# Extracting Target / Class Labels
y = cancer.target
```

```python
# print the names of the features
print(cancer.feature_names)

# print the two classes
print(cancer.target)

# print the encoding scheme for classes ['malignant'=1 'benign'=0]
print(cancer.target_names)

# check the shape of the features
print(cancer.data.shape)

#load cancer dataset as pandas dataframe
df = pd.DataFrame(cancer.data, columns=cancer.feature_names)

# Get a quick overview by printing the first 5 rows of the DataFrame:
print(df.head(5))

# Print the last 5 rows of the DataFrame:
print(df.tail())

# Return statistical description of the data in the DataFrame
df.describe()
```

```python
# Import Library for splitting data
from sklearn.model_selection import train_test_split

# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state = 50, test_size = 0.25)

# print the total number of training set (records)
print(np.shape(y_train))

# print the total number of test set (records)
print(np.shape(y_test))
```

```python
# Import Library for KNeighborsClassifier
from sklearn.neighbors import KNeighborsClassifier

# Create a training model KNeighborsClassifier
model_1 = KNeighborsClassifier()
```

```python
# Model Training, call the method fit()passing the X_train and y_train datasets as arguments.
model_1.fit(X_train, np.ravel(y_train))
```

```python
# Model Evaluation
pred=model_1.predict(X_test)

# predict the class of the first 10 lines of the X_test dataset. The return will be an array containing the estimated categories.
pred[0:10]
```

```python
#Use methods  score() and confusion_matrix() to measure the performance of your model.
model_1.score(X_test, y_test)
```

```python
#Use method confusion_matrix() to measure the performance of your model.
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, model_1.predict(X_test))
```

```python
dump(model_1, 'cancer-classifier.dmp')
```
